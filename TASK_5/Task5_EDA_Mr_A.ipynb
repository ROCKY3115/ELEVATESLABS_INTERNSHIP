**Dataset:** Titanic Dataset

**Objective:** Perform exploratory data analysis using Pandas, Matplotlib, and Seaborn to identify trends, relationships, and insights.


```
# Install (Colab only)
# !pip install seaborn pandas matplotlib --quiet

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Dataset Shape:", df.shape)
print(df.info())
print(df.head())
print("\nMissing Values:\n", df.isnull().sum())
print("\nStatistical Summary:\n", df.describe())

# Survival Count
plt.figure(figsize=(6,4))
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()
print("Observation: More passengers did not survive than survived. Approximately 38% survived while about 62% did not.")

# Age Distribution
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], kde=True, bins=30)
plt.title("Age Distribution")
plt.show()
print("Observation: Most passengers were between 20 and 40 years old. The age distribution is slightly right-skewed.")

# Survival by Gender
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
print("Observation: Female passengers had a much higher survival rate than male passengers.")

# Age vs Passenger Class
plt.figure(figsize=(6,4))
sns.boxplot(x="Pclass", y="Age", data=df)
plt.title("Age vs Passenger Class")
plt.show()
print("Observation: First-class passengers were generally older, while third-class passengers were younger and more varied in age.")

# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
print("Observation: Fare has a positive correlation with survival, whereas passenger class has a negative correlation with survival.")

print("\\nFinal Summary")
print("- Female passengers had significantly higher survival rates.")
print("- First-class passengers had the highest survival probability.")
print("- Higher ticket fares were associated with better survival.")
print("- Younger passengers were more common in third class.")
print("- Missing values exist in Age, Cabin, and Embarked columns and should be handled before modeling.")
```

## Interview Questions

1. **What is EDA?**  
EDA is the process of exploring, cleaning, and visualizing data to discover patterns and anomalies.

2. **Which plots check correlation?**  
Correlation Heatmap and Pairplot.

3. **How do you handle skewed data?**  
Log transformation, square-root transformation, or outlier treatment.

4. **How do you detect multicollinearity?**  
Using a correlation matrix or Variance Inflation Factor (VIF).

5. **What are univariate, bivariate, and multivariate analyses?**  
- Univariate: One variable.
- Bivariate: Two variables.
- Multivariate: More than two variables.

6. **Difference between Heatmap and Pairplot?**  
Heatmap shows correlation coefficients; Pairplot shows pairwise relationships visually.

7. **How do you summarize insights?**  
Highlight key trends, correlations, anomalies, and business-relevant conclusions.
